# Task 2: Data Preprocessing
###   1.	Encode categorical variables if necessary.
###   2.	Split the dataset into training and testing sets.

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

### Encode Categorical Variables (if any)

```python
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
if len(cat_cols) > 0:
    print("\n🔹 Encoding categorical columns:", list(cat_cols))
    le = LabelEncoder()
    for col in cat_cols:
        df[col] = le.fit_transform(df[col])
else:
    print(" No categorical columns found. Skipping encoding step.")
```

### Identify Target (Response) Column
### Try to find the target column automatically

```python
target_candidates = [col for col in df.columns if 'response' in col.lower()]
```

```python
if target_candidates:
    target_col = target_candidates[0]
    print(f"\n Target column detected automatically: '{target_col}'")
else:
    raise ValueError(" No column containing 'response' found. Please check your dataset.")
```

### Separate features (X) and target (y)

```python
X = df.drop(columns=[target_col])  # Independent features
y = df[target_col]                  # Target variable
```

### Feature Scaling (optional but recommended for SVM)

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### Split into Training and Testing Sets

#### 80% for training, 20% for testing

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

### Display Results

```python
print("\n Data Preprocessing Completed Successfully!")
print("Training set size:", X_train.shape)
print("Testing set size:", X_test.shape)
print("\nFeature Columns:", list(X.columns))
print("\nTarget Distribution in Train Set:\n", y_train.value_counts(normalize=True))
```

# Task 4: SVM Implementation
#### 1.	Implement a basic SVM classifier using Python libraries like scikit-learn.
###  2.	Train the SVM model on the training data.
###  3.	Evaluate model performance on the testing data using appropriate metrics (e.g., accuracy, precision, recall, F1-score).

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

### Separate features and target

```python
X = df.drop(columns=[target_col])
y = df[target_col]
```

#### Standardize features for better SVM performance

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### Split dataset into train and test sets (80/20)

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
X_train.shape[0]
```

```python
X_test.shape[0]
```

#### Implement and Train the SVM Model
#### Initialize a basic SVM model with the default RBF kernel

```python
svm_model = SVC(kernel='rbf', random_state=42)
```

#### Train the SVM model

```python
svm_model.fit(X_train, y_train)
```

### Evaluate Model Performance
### Predict responses on the test data

```python
y_pred = svm_model.predict(X_test)
```

#### Compute performance metrics

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=0)
recall = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)
```

```python
accuracy
```

```python
precision
```

```python
recall
```

```python
f1
```

#### Detailed classification report

```python
classification_report(y_test, y_pred)
```

#### Confusion Matrix Visualization

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples', cbar=False)
plt.title("Confusion Matrix - SVM Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# Task 5: Visualization of SVM Results
### 1.	Visualize classification results on the testing data.

```python
from sklearn.svm import SVC
```

#### Separate features and target

```python
X = df.drop(columns=[target_col])
y = df[target_col]
```

#### Scale features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### Split into train/test sets

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_train, y_train)
```

#### Make Predictions on Test Data

```python
y_pred = svm_model.predict(X_test)
```

#### Add predictions to a DataFrame for visualization

```python
results_df = pd.DataFrame(X_test, columns=X.columns)
results_df['Actual'] = y_test.values
results_df['Predicted'] = y_pred
```

### Visualize Classification Results
### Step 1: Confusion Matrix (visual summary)

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Confusion Matrix - SVM Results")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

#### Step 2: Scatter plot (first 2 features only, for visualization)

```python
if X.shape[1] >= 2:
    plt.figure(figsize=(7,5))
    sns.scatterplot(
        x=X_test[:, 0], y=X_test[:, 1],
        hue=y_pred, style=y_test, palette='Set2', s=70
    )
    plt.title("SVM Classification Results (Test Data)")
    plt.xlabel(X.columns[0])
    plt.ylabel(X.columns[1])
    plt.legend(title="Predicted vs Actual")
    plt.show()
else:
    print(" Not enough numeric features to visualize 2D scatter plot.")
```

#### Step 3: Compare Actual vs Predicted values

```python
comparison_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
plt.figure(figsize=(6,4))
sns.countplot(x='Actual', hue='Predicted', data=comparison_df, palette='pastel')
plt.title("Comparison of Actual vs Predicted Classes")
plt.xlabel("Actual Class")
plt.ylabel("Count")
plt.show()
```

# Task 6: Parameter Tuning and Optimization
#### 1.	Experiment with different SVM hyperparameters (e.g., kernel type, regularization parameter) to optimize performance.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

#### Define range of parameters to test

```python
param_grid = {
    'C': [0.1, 1, 10, 50],           # Regularization strength
    'kernel': ['linear', 'poly', 'rbf'],  # Kernel types
    'gamma': ['scale', 'auto']       # Kernel coefficient
}
```

#### Initialize the base SVM model

```python
svm_model = SVC(random_state=42)
```

### Perform Grid Search Cross Validation

```python
grid_search = GridSearchCV(
    estimator=svm_model,
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,          # 5-fold cross-validation
    n_jobs=-1,     # Use all CPU cores
    verbose=1      # Print progress
)
```

### Fit the grid search to the training data

```python
grid_search.fit(X_train, y_train)
```

#### Display Best Parameters and Scores

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

### Retrieve the best estimator

```python
best_svm = grid_search.best_estimator_
```

```python
best_svm
```

#### Evaluate Tuned Model on Test Data

```python
y_pred = best_svm.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
classification_report(y_test, y_pred)
```

### Visualize Confusion Matrix for Tuned Model

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', cbar=False)
plt.title("Confusion Matrix - Tuned SVM Model")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
