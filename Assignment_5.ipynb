## Data Cleannig

```python
for c in df.columns:
    if df[c].dtype == 'object':
        try:
            df[c] = pd.to_numeric(df[c].str.strip(), errors='coerce')
            print(f"Converted column '{c}' to numeric.")
        except Exception:
            pass
```

```python
df.dtypes
```

## Handle Missing

```python

for c in df.select_dtypes(include=[np.number]).columns:
    if df[c].isna().sum() > 0:
        df[c].fillna(df[c].median(), inplace=True)
        print(f"Filled missing values in {c} with median.")
```

```python
df.isnull().sum()
```

##  Histograms

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
df[numeric_cols].hist(bins=20, figsize=(10, 10))
plt.suptitle("Histograms of numeric features")
plt.show()
```

## # Correlation heatmap

```python
# Correlation heatmap
corr = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## Pairplot Sample

```python
sns.pairplot(df[numeric_cols[:6]].sample(min(500, len(df)), random_state=42))
plt.show()
```

## Save CleanFile

```python
cleaned_path = "Cardiotocographic_cleaned_5.csv"
df.to_csv(cleaned_path, index=False)
print(f"Cleaned dataset saved to: {cleaned_path}")
```
