### Removing outliers:
   #### Removing outliers:

```python
Q1_volume=np.percentile(df["Volume"],25)
Q3_volume=np.percentile(df["Volume"],75)
IQR_volume=Q3_volume-Q1_volume
IQR_volume
```

```python
#upper whisker length 
UW_volume=Q3_volume+(1.5*IQR_volume)
UW_volume
```

```python
#no of outliers in volume
print("no of outliers:",len(df[df["Volume"]>UW_volume]))
```

```python
df[df["Volume"]>UW_volume]
```

```python
df.loc[df["Volume"] > UW_volume, "Volume"] =UW_volume
```

```python
df["Volume"].describe()
```

```python
without_outliers_volume=df["Volume"].plot(kind='box',vert=False,title='Box plot for Volume after removing Outliers')
without_outliers_volume
```
